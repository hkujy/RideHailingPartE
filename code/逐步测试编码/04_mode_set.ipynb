```python
def get_mode_set(paths, theta_1, theta_2):
    """
    generate a mode set for the netwerork and calculate the mode choice probability.
    Args:
        paths ([pd.DataFrame]): [path set]
        theta_2 ([float]): [logit coefficient]

    Returns:
        [pd.DataFrame]: [a mode set inclding the mode choice probability]
    """

    # computing the log-sum term
    def get_log_sum(x):
        return (1 / theta_1) * math.log(-sum(x))
    
    mode_set = pd.DataFrame(paths.groupby(by=['mode'], as_index=False)['path_exp'].apply(get_log_sum))
    mode_set.columns =['mode', 'path_logsum']

    # computing the negative unity term
    mode_set['exp_logsum'] = -theta_2 * mode_set['path_logsum']

    # computing the mode probability
    mode_set['mode_prob'] = mode_set['exp_logsum'].apply(math.exp) / mode_set['exp_logsum'].apply(math.exp).sum()
    net_dic ={'P&R':'multi_mode',
               'R&T':'multi_mode',
               'auto': 'single_mode',
               'transit':'single_mode'}
    mode_set['net_level'] = mode_set['mode'].map(net_dic)

    return mode_set
```

```python
def get_net_level(mode_set, theta_2, theta_3):
    
    # computing the log-sum term for mode
    def get_log_sum(x):
        return (1 / theta_2) * math.log(-sum(x))
    
    net_set = pd.DataFrame(mode_set.groupby(by=['net_level'], as_index=False)['exp_logsum'].apply(get_log_sum))
    net_set.columns =['net_level', 'mode_logsum']

    # computing the exponential term
    net_set['exp_logsum'] = -theta_3 * net_set['mode_logsum']

    # computing the network probability
    net_set['net_prob'] = net_set['exp_logsum'].apply(math.exp) / net_set['exp_logsum'].apply(math.exp).sum()
    
    return net_set
```
